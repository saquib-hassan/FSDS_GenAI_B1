## 2. Count Word Frequency in Customer Reviews

üíº Real-world use: Analyzing reviews or comments in text analytics.

Task:

Given a review text, count how many times each word appears (ignore case).

Example:

Input: "This product is good and this service is excellent"

Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

# Solution

```python
feedback = input().lower()

D = {}

f = feedback.split()

for word in f:
    if word in D:
        D[word] += 1
    else:
        D[word] = 1
  
for word, count in D.items():
    print(word, count)
```

## 3. Password Strength Checker

üíº Real-world use: Used in user authentication systems.

Task:

Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one special character.

Example:

Input: "Hello@123"

Output: Strong Password

## Solution

```python
password = input("Plz enter your password: ")
if password.isupper and password.islower and password.isdigit and password.isalnum == False:
    print("Strong Password")
else:
    print("Weak Password")
```

## 4. Username Generator

üíº Real-world use: Automatically create usernames for users.

Task:

Take user‚Äôs full name and generate a short username.

Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).

Example:

Input: "Mamun Malitha"

Output: mamitha57

## Solution

```python
import random
random_num = random.randint(1,100)


full_name = input("Plz enter your full name: ")
first_name,last_name = full_name.split() #unpacking them


first_part = first_name[:3]
second_part = last_name[-3:]
username = first_part+second_part+str(random_num)
    
print(username)
```

Task:

Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS 
limit).

Example:

Input: "This is a demo message."

Output: Words = 5, Characters = 23, Status = Within Limit

## Solution

```python
sentence = input()
words = sentence.split()

character_counter = 0
word_counter = 0

for i in sentence:
    character_counter += 1
# print(character_counter)

for j in words:
    word_counter += 1
# print(word_counter)

if character_counter > 160:
    print(f"Words = {word_counter}, Characters = {character_counter}, Status = Limits crossed")
else:
    print(f"Words = {word_counter}, Characters = {character_counter}, Status = Within Limit")


#Saquib Hassan
```

## 6. Remove Duplicates from Contact List

üíº Real-world use: Data cleaning in CRM or user databases.

Task:

Given a list of names (comma separated), remove duplicates.

Example:

Input: "Rahim, Karim, Rahim, Sultana"

Output: "Rahim, Karim, Sultana"

## Solution

```python
customer_names = input("Enter names: ")
comma_seperated_names = customer_names.split(',')

stripped_names = []
for name in comma_seperated_names:
    stripped_names.append(name.strip())


cleaned_names = set(stripped_names)   

result = ", ".join(cleaned_names)
print(result)


# in this program we have whitespace
#you should give input without quotation


```

## 7. Detect Suspicious Words in Text (Moderation)

üíº Real-world use: Used in chat moderation or email filtering.

Task:

If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.

Example:

Input: "This is a scam offer"

Output: Warning: Suspicious content detected

## Solution

```python
text = input().lower()
new_text = text.split()
banned_words = ["spam","scam","fake"]

for i in new_text: # i is not the index here, rather it's the word
    if i in banned_words:
        print("Warning")
```
