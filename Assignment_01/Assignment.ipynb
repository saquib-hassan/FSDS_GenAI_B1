## 1: Temperature Check 

Problem:

Write a program that takes the current temperature as input and prints:

● "It's too cold!" if below 10°C

● "It's cool outside" if between 10°C and 25°C

● "It's hot outside!" if above 25°C

# Solution

```python
current_temperature = float(input("Please type the current temparature"))
if current_temperature < 10:
    print("It's too cold!")
elif current_temperature >= 10 and current_temperature <= 25:
    print("It's cool outside")
elif current_temperature > 25:
    print("It's hot outside!")
```

# 2: ATM Withdrawal System 

Problem:

Ask the user for their account balance and the amount they want to withdraw.

● If the withdrawal amount is greater than the balance → print "Insufficient Balance.

● If the withdrawal amount is less than or equal to balance → print "Transaction Successful" and show the remaining balance.

# Solution

```python
account_balance = float(input("How much balance do you have in your account? "))
withdraw_balance = float(input("How much do you want to withdraw? "))

if withdraw_balance > account_balance:
    print("Insufficient balance")
elif withdraw_balance <= account_balance:
    print("Transaction successfull")
    remaining_balance = account_balance - withdraw_balance
    print(f"Remainig balance in your account is: {remaining_balance}")
```
