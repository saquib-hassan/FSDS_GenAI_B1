## 1: Temperature Check 

Problem:

Write a program that takes the current temperature as input and prints:

● "It's too cold!" if below 10°C

● "It's cool outside" if between 10°C and 25°C

● "It's hot outside!" if above 25°C

# Solution

```python
current_temperature = float(input("Please type the current temparature"))
if current_temperature < 10:
    print("It's too cold!")
elif current_temperature >= 10 and current_temperature <= 25:
    print("It's cool outside")
elif current_temperature > 25:
    print("It's hot outside!")
```

# 2: ATM Withdrawal System 

Problem:

Ask the user for their account balance and the amount they want to withdraw.

● If the withdrawal amount is greater than the balance → print "Insufficient Balance.

● If the withdrawal amount is less than or equal to balance → print "Transaction Successful" and show the remaining balance.

# Solution

```python
account_balance = float(input("How much balance do you have in your account? "))
withdraw_balance = float(input("How much do you want to withdraw? "))

if withdraw_balance > account_balance:
    print("Insufficient balance")
elif withdraw_balance <= account_balance:
    print("Transaction successfull")
    remaining_balance = account_balance - withdraw_balance
    print(f"Remainig balance in your account is: {remaining_balance}")
```

# 3: Student Grade Calculator

Problem: 
Input marks (0–100) and print grade based on:

● 90–100 → A+ 

● 80–89 → A 

● 70–79 → B

● 60–69 → C
 
● Below 60 → Fail

# Solution

```python
marks = float(input("Please enter your mark: "))

if marks >=90 and marks <= 100:
    print("A+")
elif marks >= 80 and marks <= 89:
    print("A")
elif marks >= 70 and marks <= 79:
    print("B")
elif marks >= 60 and marks <= 69:
    print("C")
else:
    print("Fail")
```

# 5: Restaurant Bill with Discount 

Problem: 
Take total bill amount from the user: 

● If the bill is above 1000 → apply 10% discount 

● Otherwise → no discount 

Then print the final bill amount. 

# Solution

```python
total_bill = float(input("Plz enter your total bill: "))
if total_bill > 1000:
    discounted_bill = total_bill * 0.10
    final_bill = total_bill - discounted_bill
    print(f"This is your final bill {final_bill} after 10% discount")
else:
    print("No discount")
```

# 7: Traffic Light System 

Problem: 

Ask the user to input a color (red, yellow, or green). 

Print: 

● "Stop" for red 

● "Ready to go" for yellow

● "Invalid color" otherwise

# 8: Leap Year Checker

Problem:

Take a year as input and check: 

● If it is divisible by 400 → Leap Year 

● Else if divisible by 4 but not by 100 → Leap Year 

● Otherwise → Not a Leap Year 

```python
year = int(input("Plz enter a year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")
else:
    print("Not a Leap Year") 
```
